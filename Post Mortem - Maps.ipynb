```python
## Utility functions

def reshape_data(data):
    '''
    Aggregate and reshape data
    
    Input
    -----
    (pandas object) -> (pandas object)
    
    example
    -------
    data_agg = reshape_data(data)
    
    '''
    
    data = pd.DataFrame(data.groupby(['state_name']).size())
    data.columns = ['num_conversions']
    data['state_name'] = data.index
    return data 


def generate_percentile_range(data, states, **mode):
    
    '''
    Generate percentile ranges  -> 0th, 20th, 40th, 60th, 80th, 100th
    
    (pandas object, list, **str) -> (list, pandas object)
    
    example
    -------
    
    for approaches  -> generate_percentile_range(data, states)

    for conversions - > generate_percentile_range(data_conv, states, mode='conversions')
    
    '''
    # account for case where there are no conversions / approacf for a given state
    for name, value in mode.items():
        if value is 'conversions':
            state_conv = data.to_dict()
            state_list = []
            for state in states:
                if state in state_conv['state_name']:
                    state_list.append([state, state_conv['num_conversions'][state]])
                else:
                    state_list.append([state, 0])
                    
        data = pd.DataFrame(state_list, columns=['state_name', 'num_conversions'])

    num_conv_percentile = []
    for i in range(0,101,20):
        num_conv_percentile.append(np.percentile(data['num_conversions'], i))
    return num_conv_percentile, data

 
def generate_map(data, map_topo, num_conv_percentile, file_names, col_aggregate, col_state_name, 
                 title, scale = 1000, projection = 'albersUsa'):
    '''
    Generate maps and store .json and .html generated files in current directory
    
    Inputs
    ------
    (pandas object, json, list, array, string, string, string )
    
    
    example 
    -------
    generate_map(data_percentile, state_topo, percentile, file_names, 'num_conversions', 'state_name', 'Approaches by State')
    generate_map(data_conv_percentile, state_topo, percentile_conv, file_names, 'num_conversions', 'state_name', 'Conversions by State')

    '''
    
    geo_data = [{'name': 'states',
             'url': map_topo,
             'feature': 'us_states.geo'}]

    vis = vincent.Map(data=data, geo_data=geo_data, scale=scale,
                  projection=projection, data_bind=col_aggregate, data_key=col_state_name,
                  map_key={'states': 'properties.NAME'}, brew='YlGnBu')

    vis.scales[0].type='threshold'
    vis.scales[0].domain = num_conv_percentile
    vis.legend(title=title)
    vis.to_json(file_names[0], html_out=True, html_path=file_names[1])
```

### All Approaches by State 

```python
# load us states topo json file 
# raw State topo map at  https://raw.githubusercontent.com/wrobstory/vincent_map_data/master/us_states.topo.json
# https://github.com/wrobstory/vincent_map_data/blob/master/us_states.topo.json

state_topo = 'vincent_map_data/us_states.topo.json'

file_names =['vega_state.json','vega_state.html']
data_agg = reshape_data(data)
percentile, data_percentile  = generate_percentile_range(data_agg, states)
generate_map(data_percentile, state_topo, percentile, file_names, 'num_conversions', 'state_name', 'Approaches by State')
```

### Zip Codes 

```python
def zip_buffer(dt):
    while len(str(dt)) < 5:
        dt = '0' + str(dt)
    return dt

data_zip = pd.DataFrame(data.groupby(['zipcode']).size())
data_zip['zipcode_'] = data_zip.index

data_zip['zip_buff'] = data_zip['zipcode_'].apply(zip_buffer)
data_zip.columns = ['count','zipcode_', 'zip_buff']

# Set up the geographic data to be used
zip_topo = r'vincent_map_data/us_zip.topo.json'

geo_data = [{'name': 'zips',
             'url': zip_topo,
             'feature': 'us_zip'}]

# Bind our data to the map
vis = vincent.Map(data=data_zip, geo_data=geo_data, scale=1000, projection='albersUsa',
          data_bind='count', data_key='ZCTA5CE10',
          map_key={'zips': 'properties.ZCTA5CE10'},brew='RdPu')
vis.marks[0].properties.enter.stroke_opacity = vincent.ValueRef(value=0.5)

vis.legend(title='Number of Searches')
vis.to_json('vega.json')

vis.to_json('vega_zip.json', html_out=True, html_path= 'vega_zip.html')
```
