```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.cross_validation import train_test_split
```

`score.data` holds the feature matrix which is composed of the following four features.

`Aged_Open :` This is time difference between when the email was opened and now  

`Aged_Viewed :` This is the time difference between when the email was viewed and now

`First Open :` Boolean variable

`First View :` Boolean variable

The variables used here are only a subset from the universe of available features.

The dataset has $70$ samples and $4$ features which is further divided into test and train samples. The data has about 15% positive samples (conversions) and about 85% negative samples (no conversions)

The features used here is represented as a **feature matrix** and a **Target vector**

$$
{\rm feature~matrix:~~~} {\bf X}~=~\left[
\begin{matrix}
x_{11} & x_{12} & \cdots & x_{1D}\\
x_{21} & x_{22} & \cdots & x_{2D}\\
x_{31} & x_{32} & \cdots & x_{3D}\\
\vdots & \vdots & \ddots & \vdots\\
\vdots & \vdots & \ddots & \vdots\\
x_{N1} & x_{N2} & \cdots & x_{ND}\\
\end{matrix}
\right]
$$

$$
{\rm target~vector:~~~} {\bf y}~=~ [y_1, y_2, y_3, \cdots y_N]
$$

The dataset has $70$ samples and $4$ features.

To calculate the oportunity score for a lead, we will be exploring two approaches.

- The first approach is a machine learning based approach. Here we feed the features to an algorithm to train and then do prediction. The predictions are actually a probability of beloing to the class-1 (approach ended with a conversion) in the Target. We can rescale and use as our opportunity score

Depending on which algorithm we use for this approach, we could also make this rule based

- The second approach is a points based approach. Here we assign certain points to certain activities that occur as part of the approach. In this case the featuress in the feature matrix are rescaled and sumed up row-wise.


```python
# load feature data and target labels
score_data =  pd.read_pickle('data.pkl')
score_target =  pd.read_pickle('target.pkl')
```

```python
# sample of feature data
score_data.head()
```

```python
# target label. I - indicates conversion and 0 indicates no conversion
score_target.head()
```

```python
# divide data in to train and test
X, y = score_data, score_target
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, train_size = 0.6, random_state = 30)
```

```python
# rescale features

scalerX = StandardScaler().fit(Xtrain[:,1:])
Xtrain_scaled = scalerX.transform(Xtrain[:,1:])
Xtest_scaled = scalerX.transform(Xtest[:,1:])
```

### ``Scoring Approach 1 - Machine Learning(ML-1)``

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logit = LogisticRegression()

logit.fit(Xtrain_scaled, ytrain)
y_pred =  logit.predict_proba(Xtest_scaled)
```

```python
arr1 = pd.DataFrame(y_pred[:,1]*100, columns=['Opportunity_Score'])
arr2 = pd.DataFrame(ytest, columns=['Conversion?'])
```

```python
xtest_df = pd.DataFrame(Xtest, columns=['Approach','Aged_Open','Aged_Open','First Open','First View'])
```

```python
arr3 = pd.concat([arr1, arr2], axis=1)
score_approach_1 = pd.concat([xtest_df, arr3], axis=1)
```

```python
score_approach_1
```

### ``Scoring Approach 2 - Points Based``

```python
arr4 = pd.DataFrame(np.sum(Xtest_scaled, axis=1), columns=['Opportunity_Score'])
arr2 = pd.DataFrame(ytest, columns=['Conversion?'])
```

```python
arr5 = pd.concat([arr4, arr2], axis=1)
score_approach_2 = pd.concat([xtest_df, arr5], axis=1)
```

```python
score_approach_2
```

As a sample prototype we can see that for both scoring approaches on our test data, the leads that have the highest Opportunity score also happen to have ended in a conversion (reference the Conversion? column)
