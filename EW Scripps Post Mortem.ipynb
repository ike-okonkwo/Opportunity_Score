### DNA  - City Comparison across 51 features

```python
cities = pd.read_csv('data_cities.csv')

cities_list = ['San Diego, CA','Indianapolis, IN', 'Knoxville, TN', 'Baltimore, MD', 'Ventura, CA' ] 
cities_list_indx = np.array([0,1,2,3,4])
```

```python
cities_trans = cities.ix[:,1:].transpose()
cities_trans
```

```python
def dim_reduce(data):

    city_tsne = TSNE(learning_rate=100, n_components = 8).fit_transform(data)
    city_pca = PCA(n_components=10).fit_transform(data)
    city_mds = MDS().fit_transform(data)
    
    return city_tsne, city_pca, city_mds
```

### Radar Chart

```python
city_tsne_radar, _,_ = dim_reduce(cities_trans)
```

### Days Before Christmas

```python
### time data 
time_data = converted_leads_merge[['utc_conversion','ew_scripps_bool']]
time_data['month'] = 0
time_data['day'] = 0
time_data['diff'] = 0


# fill in month and dat columns
for i in range(len(time_data)):
    try:
        dt = parse(time_data['utc_conversion'].ix[i])
        time_data['month'][i]  = dt.month
        time_data['day'][i]  = dt.day
    except:
        time_data['month'][i] = 12
        time_data['day'][i] = 5
        
# fill in diff column

for i in range(len(time_data)):
    if time_data['month'].ix[i] == 12:
        time_data['diff'][i] = 25 - time_data['day'].ix[i]
    elif time_data['month'].ix[i] == 11:
        time_data['diff'][i] = (30 - time_data['day'].ix[i]) + 25
       
        
```
